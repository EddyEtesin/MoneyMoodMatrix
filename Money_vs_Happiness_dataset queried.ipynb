```python
##Do countries with higher Freedom to make life choices score better on the Life Ladder?**
```

```python
freedom_vs_life = df.groupby("Country").agg(
    avg_freedom=('Freedom to make life choices', 'mean'),
    avg_life_ladder=('Life Ladder', 'mean'))
freedom_vs_life 
```

```python
freedom_vs_life.plot(kind = "scatter", x = "avg_freedom", y = "avg_life_ladder")
```

```python
##**How does Generosity vary between countries with high and low happiness scores?**
```

```python
generosity_v_happiness = df.groupby("Country").agg(avg_generosity = ('Generosity', 'mean'), avg_happiness = ("Life Ladder", "mean"))
```

```python
generosity_v_happiness.plot(kind = "scatter", x = "avg_generosity", y = "avg_happiness")
```

```python
##**Which countries have the highest perceptions of corruption, and how does it affect happiness?**
```

```python
##Are countries with older median ages generally happier or less happy?*
```

```python
df.head()
countries_mid_age_V_Happiness = df.groupby("Country").agg(median_age = ("Median Age", 'mean'), happiness_level = ("Life Ladder", "mean")).sort_values(by = "median_age", ascending = False)
```

```python
countries_mid_age_V_Happiness['median_age'].corr(countries_mid_age_V_Happiness['happiness_level'])
```

```python
"""From the correlation, Countries with higher median ages are usually more happy"""
```

```python
sns.lmplot(x='median_age', y='happiness_level', data=countries_mid_age_V_Happiness)
plt.xlabel("Median Age")
plt.ylabel("Happiness Level")
```
